- Write a Python method **crawl(url)**<br>
    - It shall send a GET request to url to retrieve the HTML content (use requests package)
    - It shall parse the HTML content and extract the text (use bs4 (beau1ful soup) package).
    - Furthermore, it shall extract all links on the page and provide them as a list
    - Return values: text, links
    - For tes1ng: Crawl (for example) a single Wikipedia ar1cle you like.

```python
import requests
from bs4 import BeautifulSoup

def crawl(url):
    # Get the requests library version
    requests_version = requests.__version__
    # Build the custom User-Agent 
    # To prevent 403 Error when accessing Wikipedia
    user_agent = f"MyBot/0.0 (https://example.org/mybot/; mybot@example.org) requests/{requests_version}"
    headers = {
        "User-Agent": user_agent
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    links = []  

    anchorLinks = soup.find_all("a", href=True)
    for link in anchorLinks:
       links.append(link.get("href"))
       #links = [a.get('href') for a in soup.find_all('a', href=True)]
   
    return [soup.get_text(), links]

#url = 'https://www.oth-aw.de/'
url = "https://en.wikipedia.org/wiki/Amberg"


print(f"Text:\n {crawl(url)[0]}")

print(f"Links:\n {crawl(url)[1]}")
```
